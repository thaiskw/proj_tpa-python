```python
class Fila:
    def __init__(self):
        self.fila = []
    
    def enfileirar(self, item):
        self.fila.append(item)
        print(f"{item} chegou no caixa.")

    def vazio(self):
        return len(self.fila) == 0
    
    def desinfileirar(self):
        if not self.vazio():
            item = self.fila.pop(0)
            print(f"{item} foi atendido no caixa.")
            return item
        else:
            print("A caixa está vazia.")
            return None
        
    def tamanho(self):
        if len(self.fila) == 0:
            print("O caixa está livre.")
        else:
            print(f"Tem {len(self.fila)} pessoas na fila.")

caixa_rapido = Fila()
caixa_idosos = Fila()
caixa_geral = Fila()

def fila(nome, idade, unidades, caixa_rapido, caixa_idosos, caixa_geral):
    if unidades <= 20:
        caixa_rapido.enfileirar(nome)
    elif idade >= 60:
        caixa_idosos.enfileirar(nome)
    else:
        caixa_geral.enfileirar(nome)

pessoas = [
    ("Thainá", 16, 15),
    ("Nicole", 15, 5),
    ("Lucas", 40, 30),
    ("Pedro", 60, 18),
    ("Gustavo", 70, 20),
    ("Giovanna", 30, 10),
    ("Anna", 12, 8),
    ("Fernanda", 33, 50),
    ("Fabiana", 30, 12),
    ("Isabella", 24, 2)
]

for pessoa in pessoas:
    nome, idade, unidades = pessoa
    fila(nome, idade, unidades, caixa_rapido, caixa_idosos, caixa_geral)

print("\nFila Caixa Rápido\n")
caixa_rapido.tamanho()
while not caixa_rapido.vazio():
    caixa_rapido.desinfileirar()

print("\nFila Caixa para Idosos\n")
caixa_idosos.tamanho()
while not caixa_idosos.vazio():
    caixa_idosos.desinfileirar()

print("\nFila Caixa Geral\n")
caixa_geral.tamanho()
while not caixa_geral.vazio():
    caixa_geral.desinfileirar()
```
